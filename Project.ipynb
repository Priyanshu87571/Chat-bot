```python
import gradio as gr
from transformers import pipeline

# Load Granite pipeline
pipe = pipeline("text-generation", model="ibm-granite/granite-3.3-2b-instruct")

# Chatbot function
def chat(user_message, history, role, max_tokens):
    # Build conversation context
    messages = [{"role": "system", "content": role}]
    for user, bot in history:
        messages.append({"role": "user", "content": user})
        messages.append({"role": "assistant", "content": bot})
    messages.append({"role": "user", "content": user_message})

    # Generate response
    response = pipe(messages, max_new_tokens=max_tokens)
    bot_message = response[0]["generated_text"][-1]["content"]

    # Add to history
    history.append((user_message, bot_message))
    return history, history

# Clear history function
def clear_chat():
    return [], []

# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🤖 Granite Chatbot (Enhanced GPT-like)")

    with gr.Row():
        role = gr.Dropdown(
            ["You are a helpful assistant.",
             "You are a strict teacher.",
             "You are a funny friend.",
             "You are a coding expert."],
            value="You are a helpful assistant.",
            label="🤹 Choose Bot Personality"
        )
        max_tokens = gr.Slider(50, 500, value=200, step=50, label="Max Tokens")

    chatbot = gr.Chatbot(height=500, render_markdown=True)
    msg = gr.Textbox(placeholder="Type your message here...", label="💬 Your Message")

    with gr.Row():
        submit = gr.Button("Send")
        clear = gr.Button("🗑️ Clear Chat")

    # Wire functionality
    submit.click(chat, [msg, chatbot, role, max_tokens], [chatbot, chatbot])
    msg.submit(chat, [msg, chatbot, role, max_tokens], [chatbot, chatbot])
    clear.click(clear_chat, None, [chatbot, chatbot], queue=False)

demo.launch()
```
